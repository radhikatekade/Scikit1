```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
years_of_exp = np.array([1, 2, 3, 4, 5, 6])
print("Original array: ")
print(years_of_exp)

print("Reshaping the array: ")
years_of_exp = years_of_exp.reshape(-1, 1)
print(years_of_exp)

salaries = np.array([25000, 30000, 35000, 40000, 45000, 50000])
print("Salaries: ", salaries)
```

```python
# Linear Regression model
model = LinearRegression()
model.fit(years_of_exp, salaries)
years_to_predict = np.array([7, 8, 9]).reshape(-1, 1)
predicted_salaries = model.predict(years_to_predict)
print("predicted_salaries: ", predicted_salaries)
```

```python
from sklearn.preprocessing import LabelEncoder

categories = ['dog', 'cat', 'fish', 'dog']
label = LabelEncoder()
numeric_data = label.fit_transform(categories)
print("encoded labels: ", numeric_data)

original_data = label.inverse_transform(numeric_data)
print("back to labels: ", original_data)
```

```python
# One Hot Encoder
from sklearn.preprocessing import OneHotEncoder

colors = ['red', 'blue', 'green', 'red', 'blue']
colors = np.array(colors).reshape(-1, 1)
encoder = OneHotEncoder()
encoded_colors = encoder.fit_transform(colors)
print(encoded_colors)
encoded_colors = encoded_colors.toarray()
print(encoded_colors)
```

```python
import pandas as pd

dataset = pd.read_csv('Churn_Modelling.csv')
dataset.head()
```

```python
dataset.info()
```

```python
dataset.isnull().any()
```

```python
dataset['Age'][0] = np.nan
print(dataset.isnull().any())
print("With null entry: ")
print(dataset.head())

dataset['Age'].fillna(dataset['Age'].mean(), inplace=True)
print("Fill out the null entry: ")
print(dataset.head())
```

```python
dataset['Gender'][0] = np.nan
print(dataset.isnull().any())
print("With null entry: ")
print(dataset.head())

dataset['Gender'].fillna(dataset['Gender'].mode(), inplace=True)
print("Fill out the null entry: ")
print(dataset.head())
```
